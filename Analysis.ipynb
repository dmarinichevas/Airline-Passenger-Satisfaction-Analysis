# Airline Passenger Satisfaction analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/workspaces/Airline-Passenger-Satisfaction-Analysis/airline_passenger_satisfaction.csv')
```

```python
df.columns
```

### 1. What is the percentage of satisfied and dissatisfied customers?

```python
satisfied = df.loc[df['Satisfaction'] == 'Satisfied'].count(0)
neutral_or_dissatisfied = df.loc[df['Satisfaction'] == 'Neutral or Dissatisfied'].count(0)

plt.figure(figsize=(6,4))

labels = ['Satisfied', 'Neutral or Dissatisfied']
colors = ['g', 'r']

plt.pie([satisfied['Satisfaction'], neutral_or_dissatisfied['Satisfaction']], labels = labels, colors=colors, autopct= '%.2f %%') 

plt.title('Percentage of satisfied and dissatisfied/neutral customers')

plt.show()
```

### 6. What is the most flighable distance?

```python
print(f"Minimal flight distance {min(df['Flight Distance'])}")
print(f"Maximal flight distance {max(df['Flight Distance'])}")
```

```python
distance_gaps = []
count_of_flight = []
prev_i = 0

for i in range(100,5000,200):
  distance_gaps.append(i)
  avg = 0
  for dist in df['Flight Distance']:
    if dist < i and dist >= prev_i:
      avg += 1
  prev_i = i
  count_of_flight.append(avg)

plt.figure(figsize=(20,5))
plt.plot(distance_gaps,count_of_flight)

xticks = [i for i in range(0,5000,200)]
plt.xticks(xticks)

plt.xlabel('Distance')
plt.ylabel('Count of trips')
plt.title('Flihgt count in dependance of distance')

for i, v in zip(range(len(distance_gaps)), count_of_flight):
  plt.text(distance_gaps[i], count_of_flight[i], str(v), ha='center')

plt.show()
```

### 7. How satisfaction depends on a flight distance?

```python
satisfied = df[df['Satisfaction'] == 'Satisfied']
neutral_or_dissatisfied = df[df['Satisfaction'] == 'Neutral or Dissatisfied']
```

```python
distance_gaps_satisfied = []
count_of_flight_satisfied = []
prev_i = 0

for i in range(100,5000,200):
  distance_gaps_satisfied.append(i)
  avg = 0
  for dist in satisfied['Flight Distance']:
    if dist < i and dist >= prev_i:
      avg += 1
  prev_i = i
  count_of_flight_satisfied.append(avg)


distance_gaps_neutral_or_dissatisfied = []
count_of_flight_neutral_or_dissatisfied = []
prev_i = 0

for i in range(100,5000,200):
  distance_gaps_neutral_or_dissatisfied.append(i)
  avg = 0
  for dist in neutral_or_dissatisfied['Flight Distance']:
    if dist < i and dist >= prev_i:
      avg += 1
  prev_i = i
  count_of_flight_neutral_or_dissatisfied.append(avg)



plt.figure(figsize=(20,5))
plt.plot(distance_gaps_satisfied,count_of_flight_satisfied, 'g', label = 'Satisfied custumers')
plt.plot(distance_gaps_neutral_or_dissatisfied,count_of_flight_neutral_or_dissatisfied, 'r', label = 'Neutral or Dissatisfied custumers')


xticks = [i for i in range(0,5000,200)]
plt.xticks(xticks)

plt.xlabel('Distance')
plt.ylabel('Count of trips')
plt.title('Flihgt count in dependance of distance')

for i, v in zip(range(len(distance_gaps_satisfied)), count_of_flight_satisfied):
  plt.text(distance_gaps_satisfied[i], count_of_flight_satisfied[i], str(v), ha='center')

for i, v in zip(range(len(distance_gaps_neutral_or_dissatisfied)), count_of_flight_neutral_or_dissatisfied):
  plt.text(distance_gaps_neutral_or_dissatisfied[i], count_of_flight_neutral_or_dissatisfied[i], str(v), ha='center')

plt.legend()

plt.show()
```

### 8. What is the overall average of all satisfaction points?

```python
satisfaction_points = df[['Ease of Online Booking',
       'Check-in Service', 'Online Boarding', 'Gate Location',
       'On-board Service', 'Seat Comfort', 'Leg Room Service', 'Cleanliness',
       'Food and Drink', 'In-flight Service', 'In-flight Wifi Service',
       'In-flight Entertainment', 'Baggage Handling']]

sat_points = []
avg_values = []

for point in satisfaction_points:
  avg_val = []
  for val in satisfaction_points[point]:
    avg_val.append(val)
  value = round((sum(avg_val)/len(avg_val)),2)
  avg_values.append(value)

for point in satisfaction_points:
  sat_points.append(point)

plt.barh(sat_points, avg_values, color = 'cyan')
plt.title('Raiting of satisfaction points (0-5)')

for i, value in enumerate(avg_values):
    plt.text(value, i, str(value))



plt.show()
```
